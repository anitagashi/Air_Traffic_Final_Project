### Test & Train Split 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# Select the features (independent variables)
X = airport.drop('adjusted_passenger_count', axis=1)

# Select the target variable
y = airport['adjusted_passenger_count']
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
X_num
```

```python
X_cat
```

### Encode Categorical

```python
# Encode categorical variables
encoder = OneHotEncoder()
X_cat_encoded = encoder.fit_transform(X_cat)
```

```python
X_cat_encoded = pd.DataFrame(encoder.fit_transform(X_cat).toarray(), columns=encoder.get_feature_names_out(X_cat.columns))
```

```python
X_cat_encoded
```

### Scale the Numerical 

```python
scaler = MinMaxScaler()
X_num_scaled = pd.DataFrame(scaler.fit_transform(X_num), columns=X_num.columns)
```

```python
X_num_scaled
```

### Concat X_cat_encoded and X_num_scaled into X_normalized

```python
X_normalized = pd.concat([X_cat_encoded, X_num_scaled], axis=1)
```

### Train and Test split on X_normalized

```python
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)
```

### Initialize the regression models

```python
linear_reg = LinearRegression(n_jobs=-1)
decision_tree_reg = DecisionTreeRegressor()
knn_reg = KNeighborsRegressor(n_jobs=-1)
random_forest_reg = RandomForestRegressor(n_jobs=-1, random_state=10)
```

```python
# Train the regression models
linear_reg.fit(X_train, y_train)
```

```python
decision_tree_reg.fit(X_train, y_train)
```

```python
knn_reg.fit(X_train, y_train)
```

```python
random_forest_reg.fit(X_train, y_train)
```

### Make predictions on the testing set

```python
linear_reg_preds = linear_reg.predict(X_test)
decision_tree_reg_preds = decision_tree_reg.predict(X_test)
knn_reg_preds = knn_reg.predict(X_test)
random_forest_reg_preds = random_forest_reg.predict(X_test)
```

### Evaluate the regression models

```python
linear_reg_mse = mean_squared_error(y_test, linear_reg_preds)
linear_reg_rmse = mean_squared_error(y_test, linear_reg_preds, squared=False)
linear_reg_mae = mean_absolute_error(y_test, linear_reg_preds)
linear_reg_r2 = r2_score(y_test, linear_reg_preds)
```

```python
dt_reg_mse = mean_squared_error(y_test, decision_tree_reg_preds)
dt_reg_rmse = mean_squared_error(y_test, decision_tree_reg_preds, squared=False)
dt_reg_mae = mean_absolute_error(y_test, decision_tree_reg_preds)
dt_reg_r2 = r2_score(y_test, decision_tree_reg_preds)
```

```python
knn_reg_mse = mean_squared_error(y_test, knn_reg_preds)
knn_reg_rmse = mean_squared_error(y_test, knn_reg_preds, squared=False)
knn_reg_mae = mean_absolute_error(y_test, knn_reg_preds)
knn_reg_r2 = r2_score(y_test, knn_reg_preds)
```

```python
rf_reg_mse = mean_squared_error(y_test, random_forest_reg_preds)
rf_reg_rmse = mean_squared_error(y_test, random_forest_reg_preds, squared=False)
rf_reg_mae = mean_absolute_error(y_test, random_forest_reg_preds)
rf_reg_r2 = r2_score(y_test, random_forest_reg_preds)
```

```python
print("Linear Regression:")
print("MSE:", linear_reg_mse)
print("RMSE:", linear_reg_rmse)
print("MAE:", linear_reg_mae)
print("R-squared:", linear_reg_r2)
print()
```

```python
print("Decision Tree Regression:")
print("MSE:", dt_reg_mse)
print("RMSE:", dt_reg_rmse)
print("MAE:", dt_reg_mae)
print("R-squared:", dt_reg_r2)
print()
```

```python
print("K-Nearest Neighbors Regression:")
print("MSE:", knn_reg_mse)
print("RMSE:", knn_reg_rmse)
print("MAE:", knn_reg_mae)
print("R-squared:", knn_reg_r2)
```

```python
print("Random Forest Regression:")
print("MSE:", random_forest_reg_preds)
print("RMSE:", random_forest_reg_preds)
print("MAE:", random_forest_reg_preds)
print("R-squared:", random_forest_reg_preds)
```

### Testing a Hypothesis 

```python
from scipy import stats



# Null hypothesis (H0): The adjusted_passenger_count in 2015 is equal to 500000000
# Alternative hypothesis (H1): The adjusted_passenger_count in 2015 is not equal to 500000000

# Set the significance level (alpha)
alpha = 0.05

# Perform the t-test
t_statistic, p_value = stats.ttest_1samp(airport['adjusted_passenger_count'], 500000000)

# Compare the p-value with the significance level

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

# Print the t-statistic and p-value
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```
